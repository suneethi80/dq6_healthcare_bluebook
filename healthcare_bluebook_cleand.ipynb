Function to select only rows where healthcare provider primary taxonomy switch is Y and get the corresponding 
taxonomy code

```python
def switch (taxo_switch):
    if taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_1'] =="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_1']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_2']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_2']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_3']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_3']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_4']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_4']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_5']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_5']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_6']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_6']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_7']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_7']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_8']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_8']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_9']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_9']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_10']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_10']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_11']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_11']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_12']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_12']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_13']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_13']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_14']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_14']
    elif taxo_switch['Healthcare_Provider_Primary_Taxonomy_Switch_15']=="Y":
        return taxo_switch['Healthcare_Provider_Taxonomy_Code_15']
    else:
        return 'null'
```

```python
npi_info['taxonomy']=npi_info.apply(switch, axis=1)
```
